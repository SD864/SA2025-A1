```python

train_df = train_df.drop(columns=["Unnamed: 0"], errors='ignore')
test_df = test_df.drop(columns=["Unnamed: 0"], errors='ignore')

X_train_raw = train_df.drop(columns=["ID", "class"])
X_test_raw = test_df.drop(columns=["ID"])
y_train = train_df["class"]
test_ids = test_df["ID"]

# Impute missing NDVI values
imputer = SimpleImputer(strategy="mean")
X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train_raw), columns=X_train_raw.columns)
X_test_imputed = pd.DataFrame(imputer.transform(X_test_raw), columns=X_test_raw.columns)

def extract_time_series_features(df):
    features = pd.DataFrame()
    features["ndvi_mean"] = df.mean(axis=1)
    features["ndvi_std"] = df.std(axis=1)
    features["ndvi_max"] = df.max(axis=1)
    features["ndvi_min"] = df.min(axis=1)
    features["ndvi_range"] = features["ndvi_max"] - features["ndvi_min"]
    features["ndvi_median"] = df.median(axis=1)
    features["ndvi_skew"] = df.skew(axis=1)
    features["ndvi_early_avg"] = df.iloc[:, :9].mean(axis=1)   # Early season
    features["ndvi_mid_avg"] = df.iloc[:, 9:18].mean(axis=1)  # Mid season
    features["ndvi_late_avg"] = df.iloc[:, 18:].mean(axis=1)  # Late season
    features["ndvi_seasonal_diff"] = features["ndvi_late_avg"] - features["ndvi_early_avg"]
    return features

X_train_feat = extract_time_series_features(X_train_imputed)
X_test_feat = extract_time_series_features(X_test_imputed)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_feat)
X_test_scaled = scaler.transform(X_test_feat)

model = LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='multinomial')
cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy with Feature Engineering: {np.mean(cv_scores):.4f}")

# Fit model on full training set
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)
submission_df = pd.DataFrame({
    "ID": test_ids,
    "class": y_pred
})
submission_df.to_csv("submission.csv", index=False)
print("Enhanced submission saved to 'submission.csv'")
```
